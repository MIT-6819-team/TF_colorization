```python
def distance(x0,x1):
  return np.sqrt( (x0[0]-x1[0])**2 + (x0[1] - x1[1])**2 )
```

```python
def top5(color):
  distances = []

  for i in range(len(quantized)):
    distances.append( (distance(color,quantized[str(i)]),i) )

  a = sorted(distances)[:5]
  return (a[0][1], a[1][1], a[2][1], a[3][1], a[4][1])
```
